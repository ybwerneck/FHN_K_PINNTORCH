```python
import numpy as np
import matplotlib.pyplot as plt
#+( 0.16 + 0.3*iapp)
def integrate_neuron(u_i, v_i, iapp, DT):
    u_new = u_i + DT * 10*((iapp + 1)*(u_i*(u_i-0.4)*(1-u_i))-v_i );
    v_new = v_i + DT * ((u_i*0.04-0.16*v_i));
    
    return u_new, v_new

# Initial values
u_initial = 0.7
v_initial = 0.
iapp = 0
DT = 0.01
num_steps = int(20/DT)

# Arrays to store values
u_values = np.zeros(num_steps)
v_values = np.zeros(num_steps)

# Perform integration
u = u_initial
v = v_initial
for i in range(num_steps):
    u_values[i] = u
    v_values[i] = v
    u, v = integrate_neuron(u, v, iapp, DT)

# Plotting
plt.plot(u_values, label='u')
plt.plot(v_values, label='v')
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.title('Simulation Results')
plt.legend()
plt.show()

# Printing
print("Final values:")
print("u_final:", u_values[-1])
print("v_final:", v_values[-1])

plt.plot(u_values,v_values)
```
